```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from scipy.stats import ttest_ind
from scipy.stats import pearsonr
```

```python
# Define the path to the Excel file
excel_file_path = "./Result Table (1).xlsx"

# Read sheet 3 (index 2) from the Excel file
mixed_df = pd.read_excel(excel_file_path, sheet_name='Mixed_ds_60sh')
```

```python
# Result:
method = 'li'
```

```python

# Assuming mixed_df is already defined and includes a 'Dataset' column

# Scatter plot for AVE CGM (daily avg) & RMSE_own
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AVE CGM (daily avg)', y=method, hue='Dataset', data=mixed_df, palette='Set2', s=50)
sns.regplot(x='AVE CGM (daily avg)', y=method, data=mixed_df, scatter=False, color='red')

# Calculate correlation coefficient and p-value
corr_coef, p_value = pearsonr(mixed_df['AVE CGM (daily avg)'], mixed_df[method])

# Annotate the plot with correlation coefficient and p-value
plt.annotate(f'Correlation coefficient: {corr_coef:.2f}\nP-value: {p_value:.2e}',
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12,
             horizontalalignment='left', verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='white'))

plt.xlabel('Daily average blood glucose per subject', fontsize=14)
plt.ylabel('Blood glucose prediction RMSE', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.title('Correlation between Daily Avg BG and BG Prediction RMSE', fontsize=16)
plt.grid(True)
# Place the legend in the lower right corner
plt.legend(loc='lower right', fontsize=12)
plt.savefig(f'supp_correlation_ave_{method}.png', bbox_inches='tight')
plt.show()

# Scatter plot for TIR ratio & RMSE_own
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TIR ratio', y=method, hue='Dataset', data=mixed_df, palette='Set2', s=50)
sns.regplot(x='TIR ratio', y=method, data=mixed_df, scatter=False, color='red')

# Calculate correlation coefficient and p-value
corr_coef, p_value = pearsonr(mixed_df['TIR ratio'], mixed_df[method])

# Annotate the plot with correlation coefficient and p-value
plt.annotate(f'Correlation coefficient: {corr_coef:.2f}\nP-value: {p_value:.2e}',
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12,
             horizontalalignment='left', verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='white'))

plt.xlabel('Daily average TIR ratio per subject', fontsize=14)
plt.ylabel('Blood glucose prediction RMSE', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.title('Correlation between Daily Avg TIR ratio and BG Prediction RMSE', fontsize=16)
plt.grid(True)
plt.legend(fontsize=12)
plt.savefig(f'supp_correlation_tir_{method}.png', bbox_inches='tight')
plt.show()

# Scatter plot for TIR ratio & RMSE_own
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Gly var (daily std)', y=method, hue='Dataset', data=mixed_df, palette='Set2', s=50)
sns.regplot(x='Gly var (daily std)', y=method, data=mixed_df, scatter=False, color='red')

# Calculate correlation coefficient and p-value
corr_coef, p_value = pearsonr(mixed_df['Gly var (daily std)'], mixed_df[method])

# Annotate the plot with correlation coefficient and p-value
plt.annotate(f'Correlation coefficient: {corr_coef:.2f}\nP-value: {p_value:.2e}',
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12,
             horizontalalignment='left', verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='white'))

plt.xlabel('Daily glycemic variability per subject', fontsize=14)
plt.ylabel('Blood glucose prediction RMSE', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.title('Correlation between Daily Glycemic Variability and BG Prediction RMSE', fontsize=16)
plt.grid(True)
plt.legend(loc='lower right', fontsize=12)
plt.savefig(f'supp_correlation_gv_{method}.png', bbox_inches='tight')
plt.show()
```

```python
filtered_df = mixed_df.dropna(subset=['HbA1C'])
# Define custom colors using the second and third colors from the Set2 palette
custom_palette = {'DiaTrend': sns.color_palette('Set2')[1], 'T1DEXI': sns.color_palette('Set2')[2]}

# Scatter plot for HbA1C & RMSE
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HbA1C', y=method, hue='Dataset', data=filtered_df, palette=custom_palette, s=50)
sns.regplot(x='HbA1C', y=method, data=filtered_df, scatter=False, color='red')

# Calculate correlation coefficient and p-value
corr_coef, p_value = pearsonr(filtered_df['HbA1C'], filtered_df[method])

# Annotate the plot with correlation coefficient and p-value
plt.annotate(f'Correlation coefficient: {corr_coef:.2f}\nP-value: {p_value:.2e}',
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12,
             horizontalalignment='left', verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='white'))

plt.xlabel('HbA1C per subject(%)', fontsize=14)
plt.ylabel('Blood glucose prediction RMSE (mg/dL)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(loc='lower right', fontsize=12)
plt.savefig(f'supp_correlation_hba1c_{method}.png', bbox_inches='tight')
plt.show()
```
