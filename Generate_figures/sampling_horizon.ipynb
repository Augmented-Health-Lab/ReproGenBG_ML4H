```python
import pandas as pd
import matplotlib.pyplot as plt
# Plotting function
def plot_rmse(df, title):
    plt.figure(figsize=(10, 6))
    for method in df['Method']:
        plt.plot(df.columns[1:], df[df['Method'] == method].iloc[0, 1:], marker='o', label=method)
    plt.xlabel('Sampling Horizon')
    plt.ylabel('RMSE')
    plt.title(title)
    plt.xticks(ticks=[6, 12, 18, 24], labels=["30min", "60min", "90min", "120min"])
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot the figures
plot_rmse(ohiot1dm_df, 'Ohiot1dm Dataset')
plot_rmse(diatrend_df, 'DiaTrend Dataset')
plot_rmse(t1dexi_df, 'T1Dexi Dataset')
```
