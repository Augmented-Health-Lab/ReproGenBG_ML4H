```python
# Sort DataFrame by AGE, then by USUBJID to ensure stable visualization
df_sorted = merged_df.sort_values(by=['AGE', 'USUBJID'])

# Define colors for male and female
color_map = {'F': 'lightcoral', 'M': 'cornflowerblue'}

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(df_sorted['USUBJID'].astype(str), df_sorted['RMSE'], 
               color=df_sorted['SEX'].map(color_map), edgecolor='black')

# Annotate bars with AGE
for bar, age in zip(bars, df_sorted['AGE']):
    height = bar.get_height()
    plt.annotate(f'{age}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords='offset points',
                 ha='center', va='bottom',
                 fontsize=10, color='black')

plt.xlabel('USUBJID')
plt.ylabel('RMSE')
plt.title('RMSE by USUBJID Ordered by Age')
plt.xticks(rotation=45)
plt.tight_layout()



legend_handles = [
    Patch(color='lightcoral', label='Female'),
    Patch(color='cornflowerblue', label='Male')
]
plt.legend(handles=legend_handles, title='Sex', loc='upper right')
# Show the plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(8, 6))
sns.boxplot(x='SEX', y='RMSE', data=merged_df, palette={'F': 'lightcoral', 'M': 'cornflowerblue'})

# Customize the plot
plt.title('Distribution of RMSE by Sex')
plt.xlabel('Sex')
plt.ylabel('RMSE')
plt.tight_layout()

# Show the plot
plt.show()
```

# Based on the Martinsson result on the mixed dataset

```python
# Define age bins and labels
age_bins = np.arange(20, 80, 10)  # From 20 to 70
age_labels = [f'{i}-{i+10}' for i in age_bins[:-1]]

# Bin the ages
merged_df['AGE_GROUP'] = pd.cut(merged_df['AGE'], bins=age_bins, labels=age_labels, right=False)

# Plotting
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='AGE_GROUP', y='RMSE', data=merged_df, palette='pastel')

# Annotate the number of subjects in each age group
age_group_counts = merged_df['AGE_GROUP'].value_counts().reindex(age_labels)
for i, label in enumerate(age_labels):
    count = age_group_counts[label]
    plt.annotate(f'n={count}', 
                 xy=(i, merged_df[merged_df['AGE_GROUP'] == label]['RMSE'].max()),  # Position above the highest point in the box
                 xytext=(0, 3),  # Offset
                 textcoords='offset points', 
                 ha='center', 
                 va='bottom')

# Customize the plot
plt.title('Distribution of RMSE by Age Group')
plt.xlabel('Age Group')
plt.ylabel('RMSE')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Define the path to the CSV file
csv_file_path = "./CM.csv"

# Load the CSV file into a DataFrame
cm_df = pd.read_csv(csv_file_path)

# Display the DataFrame
cm_df.tail(5)
```

```python
# Filter the DataFrame to include only rows with USUBJID in the given list
filtered_cm_df = cm_df[cm_df['USUBJID'].isin(USUBJID)]

# Save the filtered DataFrame to a new CSV file
filtered_cm_df.to_csv("filtered_CM.csv", index=False)

# Display the filtered DataFrame
filtered_cm_df
```

```python
# Further filter to include only rows where DRDELMTD is not NaN
filtered_cm_df = filtered_cm_df[filtered_cm_df['DRDELMTD'].notna()]

# Select only the USUBJID and DRDELMTD columns
filtered_cm_df = filtered_cm_df[['USUBJID', 'DRDELMTD']]
```

```python
# Categorize the DRDELMTD column
def categorize_drdelmtd(value):
    if 'PUMP' in value.lower():
        return 'PUMP'
    elif 'closed loop' in value.lower():
        return 'CLOSED LOOP'
    else:
        return 'MDI'

filtered_cm_df['Category'] = filtered_cm_df['DRDELMTD'].apply(categorize_drdelmtd)

# Save the filtered and categorized DataFrame to a new CSV file
filtered_cm_df.to_csv("filtered_CM_categorized.csv", index=False)

# Display the filtered and categorized DataFrame
filtered_cm_df
```

```python
# Categorize the DRDELMTD column
def categorize_drdelmtd(value):
    if 'pump' in value.lower():
        return 'PUMP'
    elif 'closed loop' in value.lower():
        return 'CLOSED LOOP'
    else:
        return 'MDI'
# Remove duplicate rows based on USUBJID and Category

filtered_cm_df['Category'] = filtered_cm_df['DRDELMTD'].apply(categorize_drdelmtd)
filtered_cm_df = filtered_cm_df.drop_duplicates(subset=['USUBJID', 'Category'])

# Reindex the DataFrame
filtered_cm_df = filtered_cm_df.reset_index(drop=True)

# Save the cleaned and reindexed DataFrame to an Excel file
filtered_cm_df.to_excel("filtered_CM_cleaned.xlsx", index=False)

```
