```python
import os
import pandas as pd

# Define the base directory path
base_dir = "C:/Users/baiyi/OneDrive/Desktop/ReproGenBG_ML4H/MartinssonAndvanDoorn/artifacts/vandoorn_diatrend_experiment_24sh/nb_future_steps_6_seed_20_/"

# List of files to process
files = [f for f in os.listdir(base_dir) if os.path.isfile(os.path.join(base_dir, f))]

# Initialize a dictionary to store the data
data = {
    "File": [],
    "Metric": [],
    "Value": []
}

# Function to determine the metric from the file name
def get_metric_from_filename(filename):
    parts = filename.split('_')
    if len(parts) == 3:
        return parts[1].upper() + "_" + parts[2].split('.')[0].upper()
    elif len(parts) == 2:
        return parts[1].split('.')[0].upper()
    return "UNKNOWN"

# Read each file and extract the value
for file in files:
    file_path = os.path.join(base_dir, file)
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            value = float(f.read().strip())
            metric = get_metric_from_filename(file)
            data["File"].append(file)
            data["Metric"].append(metric)
            data["Value"].append(value)
    except (UnicodeDecodeError, ValueError) as e:
        print(f"Skipping file {file} due to error: {e}")

# Create a DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to an Excel file
df.to_excel("./result_tables/vandoorn_diatrend_experiment_24sh_metrics.xlsx", index=False)

print("Data has been written to metrics.xlsx")
```

```python
import os
import pandas as pd

# Define the base directory path
base_dir = "C:/Users/baiyi/OneDrive/Desktop/ReproGenBG_ML4H/MartinssonAndvanDoorn/artifacts/mixed_dataset_12sh/nb_future_steps_6_seed_10_/"

# List of files to process
files = [f for f in os.listdir(base_dir) if os.path.isfile(os.path.join(base_dir, f))]

# Initialize a dictionary to store the data
data = {
    "File": [],
    "Number": [],
    "Metric": [],
    "Value": []
}

# Function to determine the number and metric from the file name
def get_number_and_metric_from_filename(filename):
    parts = filename.split('_')
    if len(parts) >= 2:
        number_part = parts[0].replace('Subject', '')
        metric_part = parts[-1].split('.')[0].upper()  # Get the part before the last extension
        return number_part, metric_part
    return "UNKNOWN", "UNKNOWN"

# Read each file and extract the value
for file in files:
    file_path = os.path.join(base_dir, file)
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            value = float(f.read().strip())
            number, metric = get_number_and_metric_from_filename(file)
            data["File"].append(file)
            data["Number"].append(number)
            data["Metric"].append(metric)
            data["Value"].append(value)
    except (UnicodeDecodeError, ValueError) as e:
        print(f"Skipping file {file} due to error: {e}")

# Create a DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to an Excel file
df.to_excel("./result_tables/mixed_dataset_12sh.xlsx", index=False)

```
