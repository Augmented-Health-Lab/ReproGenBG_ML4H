```python
import os

def replace_string_in_yaml_files(root_dir, old_string, new_string):
    """
    Replace a specific string in all .yaml files under a directory (including subdirectories).

    Args:
        root_dir (str): The root directory to search for .yaml files.
        old_string (str): The string to be replaced.
        new_string (str): The string to replace with.
    """
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".yaml"):
                file_path = os.path.join(subdir, file)
                
                # Read the file content
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                
                # Replace the old string with the new string
                updated_content = content.replace(old_string, new_string)
                
                # Write the updated content back to the file
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(updated_content)
                
                print(f"Updated: {file_path}")

# Example usage
root_directory = "./original_diatrend_experiments_60min"
old_path = "../../datasets/diatrend_subset/"
new_path = "../../../datasets/diatrend_subset/"
replace_string_in_yaml_files(root_directory, old_path, new_path)
```
