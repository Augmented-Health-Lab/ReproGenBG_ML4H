# Martinsson et al - For a dataset with a existing yaml template


```python
import os
import yaml

# Define the base directory path
base_dir = "C:/Users/baiyi/OneDrive/Desktop/ReproGenBG_ML4H/MartinssonAndvanDoorn/original_ohio_experiments_90min/"
new_dir = "C:/Users/baiyi/OneDrive/Desktop/ReproGenBG_ML4H/MartinssonAndvanDoorn/original_ohio_experiments_120min/"

# Create the new directory if it doesn't exist
os.makedirs(new_dir, exist_ok=True)

# List of YAML files to process
yaml_files = [f for f in os.listdir(base_dir) if f.endswith('.yaml')]

# Function to update the YAML content
def update_yaml_content(content):
    if 'dataset' in content and 'nb_past_steps' in content['dataset']:
        content['dataset']['nb_past_steps'] = 24
    if 'train' in content and 'artifacts_path' in content['train']:
        content['train']['artifacts_path'] = "../artifacts/martinsson_ohio_experiment_24sh/"
    return content

# Process each YAML file
for yaml_file in yaml_files:
    file_path = os.path.join(base_dir, yaml_file)
    
    # Read the YAML file
    with open(file_path, 'r') as f:
        content = yaml.safe_load(f)
    
    # Update the content
    updated_content = update_yaml_content(content)
    
    # Write the updated content to a new YAML file in the new directory
    new_file_path = os.path.join(new_dir, yaml_file)
    with open(new_file_path, 'w') as f:
        yaml.safe_dump(updated_content, f)
    
    print(f"Updated file written to {new_file_path}")
```

# Generate yaml for the mixed dataset - for Martinsson method

```python
import os
import yaml

# Define the base directory paths for the datasets
ohio_dir = "C:/Users/baiyi/OneDrive/Desktop/Modify_GenBG/OhioT1DM 2020"
diatrend_dir = "C:/Users/baiyi/OneDrive/Desktop/Modify_GenBG/modified_diatrend_subset"
t1dexi_dir = "C:/Users/baiyi/OneDrive/Desktop/Modify_GenBG/modified_t1dexi_subset/T1DEXI_cgm_processed"
new_yaml_dir = "C:/Users/baiyi/OneDrive/Desktop/ReproGenBG_ML4H/MartinssonAndvanDoorn/mixed_dataset_60min/"

# Create the new directory if it doesn't exist
os.makedirs(new_yaml_dir, exist_ok=True)

# List of files in the dataset directories
ohio_xml_files = [f for f in os.listdir(ohio_dir) if f.endswith('.xml')]
diatrend_csv_files = [f for f in os.listdir(diatrend_dir) if f.endswith('.csv')]
t1dexi_csv_files = [f for f in os.listdir(t1dexi_dir) if f.endswith('.csv')]

# Base YAML content to be used for generating new YAML files
base_yaml_content = {
    'dataset': {
        'script_path': '',
        'csv_path': '',
        'nb_past_steps': 12,
        'param_nb_future_steps': [6],
        'train_fraction': 0.0,
        'valid_fraction': 0.0,
        'test_fraction': 1.0,
        'scale': 0.01
    },
    'model': {
        'script_path': '../../models/lstm_experiment_keras.py',
        'nb_lstm_states': 256,
        'activation_function': 'exp'
    },
    'optimizer': {
        'script_path': '../../optimizers/adam_keras.py',
        'learning_rate': 1e-3
    },
    'loss_function': {
        'script_path': '../../loss_functions/nll_keras.py'
    },
    'train': {
        'script_path': '../../train/train_keras.py',
        'artifacts_path': '../../artifacts/mixed_dataset_12sh/',
        'batch_size': 1024,
        'epochs': 10000,
        'patience': 200,
        'shuffle': True,
        'param_seed': [10]
    }
}

# Function to generate YAML files
def generate_yaml_files(files, base_dir, script_path, new_yaml_dir):
    for file in files:
        # Update the script_path and csv_path with the new file path
        base_yaml_content['dataset']['script_path'] = script_path
        base_yaml_content['dataset']['csv_path'] = os.path.join(base_dir, file)
        
        # Create a new YAML file name based on the file name
        yaml_file_name = os.path.splitext(file)[0] + '_evaluation.yaml'
        new_yaml_file_path = os.path.join(new_yaml_dir, yaml_file_name)
         
        # Write the updated content to the new YAML file
        with open(new_yaml_file_path, 'w') as f:
            yaml.safe_dump(base_yaml_content, f)
        
        print(f"Generated YAML file: {new_yaml_file_path}")

# Generate YAML files for each dataset
generate_yaml_files(ohio_xml_files, ohio_dir, '../../datasets/ohio.py', new_yaml_dir)
generate_yaml_files(diatrend_csv_files, diatrend_dir, '../../datasets/diatrend.py', new_yaml_dir)
generate_yaml_files(t1dexi_csv_files, t1dexi_dir, '../../datasets/t1dexi.py', new_yaml_dir)
```
